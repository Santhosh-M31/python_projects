```python
import pandas as pd
```

```python
df = pd.read_csv("data/sample.csv")

df
```

```python
result_df = df.groupby('id').agg({
    'mean': 'mean',
    'median': 'median',
    'mode': lambda x: x.mode().iloc[0] if not x.mode().empty else None,
    'max': 'max',
    'min': 'min'
}).reset_index()

result_df
```
